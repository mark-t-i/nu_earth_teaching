```
import pandas as pd
import ternary
```

```
# Import local file
from google.colab import files
uploaded = files.upload()
```

```
# Store Dataset in a Pandas Dataframe
import io
df1 = pd.read_csv(io.BytesIO(uploaded['QFL_demo.csv']))
```

```
# Print Dataframe
print (df1)
```

```
# Set up the figure.
fig, tax = ternary.figure(scale=100)
fig.set_size_inches(10, 9)

# Plot points.
# Label axes.
tax.scatter(df1[['F', 'L', 'Q']].values)

# Axis labels. (See below for corner labels.)
fontsize = 14
offset = 0.08
tax.left_axis_label("Q %", fontsize=fontsize, offset=offset)
tax.right_axis_label("F %", fontsize=fontsize, offset=offset)
tax.bottom_axis_label("F %", fontsize=fontsize, offset=-offset)
tax.set_title("QFL Plot", fontsize=20)

# Decoration.
tax.boundary(linewidth=1)
tax.gridlines(multiple=10, color="gray")
tax.ticks(axis='lbr', linewidth=1, multiple=20)
tax.get_axes().axis('off')

tax.show()
```
